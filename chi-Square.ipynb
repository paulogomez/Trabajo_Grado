```python
import pandas as pd
date=pd.read_csv('hurtos_pasto.csv')
date
```

## colocar las variables que se relacionaran 

```python
contingency_table=pd.crosstab(date["dia"],date["barrio"])
print('contingency_table :-\n',contingency_table)
```

## vector con los datos obtenidos

```python
Observed_Values = contingency_table.values 
print("Observed Values :-\n",Observed_Values)
```

```python
import scipy.stats
b=scipy.stats.chi2_contingency(contingency_table)
Expected_Values = b[3]
print("Expected Values :-\n",Expected_Values)
```

```python
no_of_rows=len(contingency_table.iloc[0:2,0])
no_of_columns=len(contingency_table.iloc[0,0:2])
df=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",df)
```

```python
alpha=0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
print("chi-square statistic:-",chi_square_statistic)
```

```python
critical_value=chi2.ppf(q=1-alpha,df=df)
print('critical_value:',critical_value)
```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=df)
print('p-value:',p_value)
```

```python
print('Significance level: ',alpha)
print('Degree of Freedom: ',df)
print('chi-square statistic:',chi_square_statistic)
print('critical_value:',critical_value)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Rechazar H0, existe una relación entre 2 variables categóricas")
else:
    print("Retener H0, no hay relación entre 2 variables categóricas")
    
if p_value<=alpha:
    print("Rechazar H0, existe una relación entre 2 variables categóricas")
else:
    print("Retener H0, no hay relación entre 2 variables categóricas")
```
